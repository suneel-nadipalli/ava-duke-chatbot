```python
import datasets

import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/duke-qac/qaduke5_700.csv') 
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python
from datasets import DatasetDict,Dataset
```

```python
df.head()
```

```python
train_ds = Dataset.from_pandas(train)

train_ds = train_ds.remove_columns(["__index_level_0__"])

test_ds = Dataset.from_pandas(test)

test_ds = test_ds.remove_columns(["__index_level_0__"])
```

```python
train_ds
```

```python
duke_qac_ds = DatasetDict(
{
    'train': train_ds,
    'val': test_ds
})
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
duke_qac_ds.push_to_hub('suneeln-duke/duke_qac_v3')
```
